```python
import os
from glob import glob

import matplotlib.pyplot as plt
import numpy as np
import geopandas as gpd
import xarray as xr
import rioxarray as rxr
import earthpy as et
import earthpy.spatial as es
import earthpy.plot as ep
```

```python
data = et.data.get_data("cold-springs-fire")
os.chdir(os.path.join(et.io.HOME, 
                      "earth-analytics", 
                      "data"))
```

```python
lndsat_pst_fire_pth = os.path.join("cold-springs-fire",
                                      "landsat_collect",
                                      "LC080340322016072301T1-SC20180214145802",
                                      "crop")

pst_fire_pths = glob(os.path.join(lndsat_pst_fire_pth,
                                  "*band*.tif"))

pst_fire_pths.sort()
pst_fire_pths
```

```python
band_1 = rxr.open_rasterio(pst_fire_pths[0],
                           masked=True)
band_1.shape
```

```python
f, ax=plt.subplots()
band_1.plot(ax=ax, cmap="Greys_r")
ax.set_axis_off()
ax.set_title("Plot of Band 1")
plt.show()
```

```python
def open_clean_bands(band_path):
     return rxr.open_rasterio(band_path, masked=True).squeeze()
```

```python
all_bands = []
for i, aband in enumerate(pst_fire_pths):
    all_bands.append(open_clean_bands(aband))
    # Assign a band number to the new xarray object
    all_bands[i]["band"]=i+1
```

```python
lndst_pst_fire_xr = xr.concat(all_bands, dim="band") 
lndst_pst_fire_xr
```

```python
lndst_pst_fire_xr.plot.imshow(col="band",
                             col_wrap=3,
                             cmap="Greys_r")
plt.show()
```

```python
ep.plot_rgb(lndst_pst_fire_xr.values,
            rgb=[3, 2, 1],
            title="RGB Composite Image\n Post Fire Landsat Data")

plt.show()
```

```python
ep.plot_rgb(lndst_pst_fire_xr.values,
            rgb=[3, 2, 1],
            title="Landsat RGB Image\n Line Stretch Applied",
            stretch=True,
            str_clip=1)

plt.show()
```

```python
ep.plot_rgb(lndst_pst_fire_xr.values,
            rgb=[3, 2, 1],
            title="Landsat RGB Image\n Line Stretch Applied",
            stretch=True,
            str_clip=4)

plt.show()
```

```python
band_titles = ["Band 1", 
               "Blue", 
               "Green", 
               "Red",
               "NIR", 
               "Band 6", 
               "Band7"]

ep.hist(lndst_pst_fire_xr.values,
        title=band_titles)

plt.show()
```
